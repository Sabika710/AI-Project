```python
Data = df.copy()
```

```python
Data.rename(columns={'Choose your gender': 'gender'}, inplace=True)
# , 'Your current year of Study ': 'Your current year of Study'
```

```python
plt.figure(figsize=(5,5))
plt.hist(df['Age'],color='y')
plt.title("Age distribution");
```

```python
plt.figure(figsize=(6,6))
sns.countplot(data=Data, x='Your current year of Study', hue='gender')
plt.title("Students studying in particular year")
plt.show()
```

```python
plt.figure(figsize=(14, 5))
plt.title('Gender Distribution')
gender_counts = Data['gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',  explode=(0.005, 0.005), labeldistance=0.4 )
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(data=df, x='Your current year of Study', hue='Do you have Depression?')
plt.title("Students studying in particular year")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(data=df, x='Do you have Anxiety?', hue='Do you have Depression?')
plt.title("Anxiety vs Depression among students")
plt.show()
```
